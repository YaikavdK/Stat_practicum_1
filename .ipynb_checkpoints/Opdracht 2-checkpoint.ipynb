## Data inladen en voorbereiden

```python
psych = pd.read_csv('psych.csv')
```

## Individual based summary maken

Er wordt een individual based summary (IBS) gemaakt. Hierin worden de belangrijkste gegevens uit de dataset per individu samengevat. Er worden dummies gemaakt van `main_hobby`, `ad_topic` en `device_type`, omdat deze data categorisch is.

Ook is er voor gekozen om te kijken wanneer de hobbies (`main_hobby`) overeen komen met het thema van de advertentie (`ad_topic`). Daarmee kan gekeken worden of mensen met een bepaalde hobby langer blijven kijken als de advertentie over die hobby gaat. Deze worden `sports_equal`, `gaming_equal` en `fashion_equal` genoemd.

Van `skipped?` wordt de kolom `Event` gemaakt. Hiervoor geldt:
- 1: Event heeft plaatsgevonden na survival time
- 0: Deelnemer heeft hele advertentie uitgekeken
- NaN: Meetfout tijdens experiment 

**Een event neemt dus plaats als een persoon de advertentie overslaat.**

```python
psych_ibs = pd.DataFrame()

psych_ibs = pd.get_dummies(psych, columns=['ad_topic', 'main_hobby', 'device_type'], drop_first=False)
psych_ibs["fashion_equal"] = ((psych_ibs["ad_topic_fashion"] == True) &(psych_ibs["main_hobby_fashion"] == True))
psych_ibs["sports_equal"] = ((psych_ibs["ad_topic_sports"] == True) &(psych_ibs["main_hobby_sports"] == True))
psych_ibs["gaming_equal"] = ((psych_ibs["ad_topic_gaming"] == True) &(psych_ibs["main_hobby_gaming"] == True))
psych_ibs["Event"] = psych["skipped?"].map({'yes': 1, 'no': 0})
psych_ibs["Event"] = psych_ibs["Event"].fillna(0) # Meetfouten ook op 0 zetten, geen event waargenomen
psych_ibs['Censoring'] = 1 - psych_ibs["Event"]
psych_ibs = psych_ibs.rename(columns={'skip_time': 'Survival_time'})
psych_ibs.drop(columns=['skipped?'], inplace=True)

psych_ibs 
```

## CPH-model maken
Er wordt een CPH-model (Cox Proportional Hazards model) gefit met behulp van de CoxPHFitter binnen lifelines. Dit model schat hoe de verschillende factoren de kans op een event beïnvloeden.

Eerst wordt er gekeken naar alle mogelijke beschikbare factoren.

Er wordt gestreven naar een zo'n hoog mogelijke Concordance Index, in de realiteit is een index tussen 0.55 en 0.75 een goed streefpunt.

Op basis van dit model zal gekeken worden welke factoren statistische significantie aantonen.
Hierbij is $\alpha = 0.05$ aangenomen.

De AIC’s van verschillende modellen kunnen met elkaar vergeleken worden als de modellen op exact dezelfde data zijn gefit. Het meet hoe goed het model bij de data past. Hoe lager hoe beter.

```python
cph = lifelines.CoxPHFitter()
cph.fit(psych_ibs, duration_col='Survival_time', event_col='Event', formula='ad_topic_sports + ad_topic_gaming + main_hobby_sports + main_hobby_gaming + device_type_desktop + device_type_laptop + fashion_equal + sports_equal + gaming_equal + screen_brightness + age')
```

```python
cph.print_summary()
```

Hieronder zijn, van elke factor, de 95%-Confidence Interval van de coëfficiënten te zien. De factoren zijn statistisch significant als deze niet de stippellijn kruisen.

```python
cph.plot()
```

Een CPH-model maakt de aanname dat de hazards proportioneel zijn, dit betekent dat de hazard ratio niet afhankelijk is van de tijd, maar van de verhoudingen van de andere hazard ratios. Hieronder wordt gecontroleerd of het model voldoet aan deze aanname.

```python
cph.check_assumptions(psych_ibs, p_value_threshold=0.05, show_plots=False)
```

Dit model haalt een Concordance Index van `0.72` en een Partial AIC van `3597.48`. Er wordt naar gestreven om de Concordance Index verder te verhogen en de AIC verder te verlagen in een nieuw model.

De p-value van `sports_equal`, `gaming_equal` en `fashion_equal` de enige zijn die kleiner zijn dan $\alpha = 0.05$. Deze factoren zijn statistisch significant. Er wordt hieronder een nieuw model gemaakt waarin alleen de statistisch significante factoren gebruikt worden, om te kijken of de schattingen beter worden.

```python
cph2 = lifelines.CoxPHFitter()
cph2.fit(psych_ibs, duration_col='Survival_time', event_col='Event', formula='fashion_equal + sports_equal + gaming_equal')

cph2.print_summary()
```

Opnieuw zijn hieronder, van elke factor, de 95%-Confidence Interval van de coëfficiënten te zien. De factoren zijn statistisch significant als deze niet de stippellijn kruisen.

```python
cph2.plot()
```

Er wordt nog een keer gecontroleerd of de aanname van de hazard ratio klopt:

```python
cph2.check_assumptions(psych_ibs, p_value_threshold=0.05, show_plots=False)
```

## Resultaten

In het versimpelde model is te zien dat alle factoren statistisch significant zijn. De Concordance Index is afgenomen met `0.03` van `0.72` naar `0.69`. Verder is de Partial AIC afgenomen van `3597` naar `3587`.

De daling van `0.03` punten in de Concordance Index is een minimale daling en valt nog steeds binnen de gewenste range van `0.55-0.75`.

De AIC is met een waarde van 10 afgenomen, wat aangeeft dat met de vereenvoudiging dit nieuwe model de data nog steeds goed vastlegt, aangezien het model van 11 naar 3 factoren is gegaan.

Hieronder worden de significante factoren geplot. In deze grafieken kan worden afgelezen wat de kans is dat de deelnemers de advertentie wegklikken afhankelijk van of de advertentie bij hun hobby past of niet.

```python
sns.set_style('darkgrid')
cph2.plot_partial_effects_on_outcome(covariates = ['fashion_equal'], values = [False, True], plot_baseline = True, color=['#f5aa42', '#b717bd'])
plt.xlabel('seconden')
plt.ylabel('survival probability')
plt.title('Effect van gelijke hobby en ad topic (fashion) op survival times')
plt.show()
```

```python
sns.set_style('darkgrid')
cph2.plot_partial_effects_on_outcome(covariates = ['gaming_equal'], values = [False, True], plot_baseline = True, color=['#f5aa42', '#b717bd'])
plt.xlabel('seconden')
plt.ylabel('survival probability')
plt.title('Effect van gelijke hobby en ad topic (gaming) op survival times')
plt.show()
```

```python
sns.set_style('darkgrid')
cph2.plot_partial_effects_on_outcome(covariates = ['sports_equal'], values = [False, True], plot_baseline = True, color=['#f5aa42', '#b717bd'])
plt.xlabel('seconden')
plt.ylabel('survival probability')
plt.title('Effect van gelijke hobby en ad topic (sports) op survival times')
plt.show()
```

## Conclusie

De enige statistisch significante factoren die beïnvloeden of iemand een advertentie wel of niet wegklikt lijken de hobbies en het thema van de advertentie te zijn. Deelnemers klikken de advertentie minder snel weg als het thema van deze overeenkomt met hun hobby.

Door bijvoorbeeld te kijken naar de grafieken waar het 95%-CI van de coëfficiënten geplot zijn, valt te zien dat de factoren die statistisch significant ver weg liggen van de stippellijn. Verder voldoet het model aan de proportionele aanname en het model heeft een redelijk hoge Concordance Index. Dit samen zorgt voor een hoge zekerheid in de resultaten.
