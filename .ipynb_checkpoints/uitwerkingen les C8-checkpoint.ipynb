## Benodigde libraries, functies en data  
Zoals altijd beginnen we met het importeren van het een en ander. Allereerst de libraries.

```python
# importeer libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.arima.model import ARIMA
import seaborn as sns
sns.set_style('whitegrid')
```

Hieronder volgt een code chunk met een functie die deze les goed van pas komt.

```python
# functie voor plotten van time series
def plot_series(df, series_name, lags=40, diff=0, seasonal_diff=0, seasonal_period=12):
    """
    Plot a time series alongside its ACF and PACF with Bartlett bounds.
    Includes options for regular and seasonal differencing.
    
    Parameters
    ----------
    df : pandas.DataFrame
        DataFrame containing multiple time series as columns.
    series_name : str
        Column name of the series to plot.
    lags : int
        Number of lags for ACF/PACF.
    diff : int, default=0
        Number of regular differences to apply.
    seasonal_diff : int, default=0
        Number of seasonal differences to apply.
    seasonal_period : int, default=12
        Seasonal period (e.g., 12 for monthly data with yearly seasonality).
    """
    series = df[series_name]

    # Apply differencing
    for _ in range(diff):
        series = series.diff()
    for _ in range(seasonal_diff):
        series = series.diff(seasonal_period)
    series = series.dropna()

    # Compute ACF/PACF (drop lag 0)
    acf_vals = acf(series, nlags=lags, fft=False)[1:]
    pacf_vals = pacf(series, nlags=lags, method="ywm")[1:]
    lags_range = np.arange(1, len(acf_vals)+1)

    # Bartlett bounds
    n = len(series)
    conf = 1.96 / np.sqrt(n)

    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    # Time series
    axes[0].plot(series.index, series.values, color="steelblue")
    axes[0].set_title(f"Series: {series_name} (diff={diff}, seas_diff={seasonal_diff})")

    # ACF
    axes[1].stem(lags_range, acf_vals, basefmt=" ")
    axes[1].hlines([conf, -conf], xmin=0, xmax=lags, colors="red", linestyles="dashed")
    axes[1].axhline(0, color="black", linewidth=0.8)
    axes[1].set_title("ACF-plot")

    # PACF
    axes[2].stem(lags_range, pacf_vals, basefmt=" ")
    axes[2].hlines([conf, -conf], xmin=0, xmax=lags, colors="red", linestyles="dashed")
    axes[2].axhline(0, color="black", linewidth=0.8)
    axes[2].set_title("PACF-plot")

    plt.tight_layout()
    plt.show()
```

## Inferentie: wat is dat?  
De kreet 'inferentie' heb je al eens eerder gehoord in de opleiding, maar het kan geen kwaad die hier te herhalen. Een inferentie is een afleiding of gevolgtrekking. Infereren betekent dat je conclusies trekt op basis van eerdere waarnemingen. Dat doet hopelijk snel denken aan Statistiek 2, waar we het vaak hebben gehad over _statistische significantie_.  

Net als bij Statistiek 2 is time series inferentie steeds onder de aanname dat de data representatief is voor de toekomst. Dat is behoorlijk anders dan bij time series forecasting, want daar hebben modellen de vrijheid om te 'meebewegen' met de data wanneer die zich anders gaat gedragen. De aanname bij time series inference maakt ook dat het dus niet nodig is om te testen hoe het in de nabije toekomst eruitziet. Gevolg is dat een cross-validation opzet dus **niet** nodig is. Sterker nog, het is zelfs ongewenst omdat je door een eventuele split niet de volledige data gebruikt om op significantie te toetsen. En: hoe minder data, hoe minder betrouwbaar je conclusie.  

Inferentie aan de hand van een time series model, bijvoorbeeld een ARIMA, gaat op de manier die je al kent van Statistiek 2. Eerst gaan we na of het model statistisch valide is. Alleen als dat het geval is, mogen we wat zeggen over statistische significantie.

## Statistische validiteit 
Over lineaire regressie zeiden we bij Statistiek 2 dat dat alleen statistisch valide is, als aan een viertal eigenschappen is voldaan. Samen komen de vier eigenschappen neer op: het model vangt de kenmerkende eigenschappen van de data dusdanig goed op, dat wat overblijft (de residuen) ruis is waar geen statistische waarde meer uit te halen valt. We willen iets vergelijkbaars voor time series modellen. Daarvoor moeten we op zoek naar een time series die, statistisch gezien, geen meerwaarde heeft.

### White noise  
Wellicht denk je in eerste instantie bij een time series zonder statistische meerwaarde aan een constante time series. Maar die heeft juist veel meerwaarde: dat de time series constant is, vertelt ons dat de achterliggende kansvariabele dus (nagenoeg) geen variantie heeft en we zien ook gelijk wat de exacte waarde van de achterliggende kansvariabele (of eigenlijk dus: de constante) is.  

Een time series zonder statistische meerwaarde is een time series die zodanig onregelmatig is, dat daar geen interessante patronen uit te halen zijn. Een dergelijke time series noemen we _white noise_. Een time series noemen we _white noise_ als die aan de volgende drie eigenschappen voldoet:  
1. Centraliteit: de time series is gemiddeld nul  
2. Homoscedasticiteit: de variantie van de time series is constant over de hele tijd  
3. Geen autocorrelatie: de time series bevat geen significante autocorrelaties

**_Opgave 1_**  
Stel dat je een time series hebt en je wilt nagaan of de time series ook white noise is. Dan moet je dus 3 voorwaarden nagaan.  
Geef voor elke voorwaarde aan met welke methode je kan nagaan of daaraan voldaan is.  

Centraliteit: two-sided $t$-test op residuals  
Homoscedasticiteit: plot residuals over de tijd, spreiding moet niet groter/kleiner worden  
Geen autocorrelatie: bekijk ACF-plot van residuals

**_Opgave 2_**  
Met onderstaande code simuleer je een time series die white noise zou moeten zijn.  
Ga na, aan de hand van de methoden die je in Opgave 1 hebt beschreven, of de gesimuleerde time series ook echt white noise is.

```python
# simuleer white noise
sim_series = np.random.normal(loc=0, scale=0.1, size=300)
```

```python
# ga centraliteit na
from statsmodels.stats.weightstats import DescrStatsW
_, p, _ = DescrStatsW(sim_series).ttest_mean(value=0)
print(f"The p-value of the t-test for centrality is: {p:.3f}")
```

```python
# ga homoscedasticiteit na
sim_df = pd.DataFrame(sim_series)
plot_series(sim_df, 0)
```

### Statistische validiteit van een time series model  
We zijn nu zover om een statistisch valide time series model te definiren. Een time series model is _statistisch valide_ dan en slechts dan als de residuen van het model white noise zijn.  

Relatief eenvoudig, dus. We hoeven ons niet druk te maken om het splitten van de data (in welk opzicht dan ook) en hoeven van de residuen slechts 3 voorwaarden na te gaan. Toch is het een illusie om te denken dat er voor elke time series een statistisch valide model te vinden valt: zeker bij time series die sterk afhankelijk zijn van externe invloeden, is statistische validiteit lang niet altijd mogelijk. Immers, een time series model kijkt vooral naar interne patronen.

**_Opgave 3_**  
Bekijk de passagiersaantallen op Schiphol en fit daarop het best voorspellende ARIMA-model van les C7.  
Plot de residuen van dit model. Wat valt je op?

```python
model = ARIMA(nl_airports['AMS'], order=(2,0,0), seasonal_order=(2,1,1,12))
fitted = model.fit()
resid = pd.DataFrame(fitted.resid)
plot_series(resid, 0)
```

Het maakt niet uit welk ARIMA-model je fit; de eerste residuen zijn altijd heel sterk. Dat komt doordat een ARIMA-model chronologisch leert. In het begin heeft het model nog geen eerdere waarnemingen waarop teruggevallen kan worden, dus moeten daar initile schattingen gebruikt worden. Schattingen die vaak niet ergens op gebaseerd zijn. Voor het model als geheel is dat geen groot probleem; als er maar genoeg tijdstippen zijn, heeft het ARIMA-model voldoende mogelijkheid om alsnog te leren.  

Voor statistische validiteit geven die initile residuen wel een verkeerd beeld. Dat zijn residuen die niet representatief zijn voor de mate waarin het model daadwerkelijk van de data aan het leren is. Die residuen moeten dus buiten beschouwing gelaten worden. Hoeveel initile residuen buiten beschouwing gelaten moeten worden, hangt af van het gefitte model: als je een ARIMA($p,d,q$)($P,D,Q$)$_m$ hebt gefit, laat je de eerste $\max(p, q, Pm, Qm)$ residuen buiten beschouwing. Vaak is het dus de seasonality die hier de doorslag geeft.

## Inferentie  
Het verfijnen van een time series model om tot een statistisch valide model te komen, is niet altijd een voor de hand liggend proces. En, soms is dat gewoon niet mogelijk. Maar in de gevallen dat dat wel kan en je tot een statistisch valide model gekomen bent, is het gelukkig niet heel moeilijk om de daadwerkelijke inferentie uit te voeren.

**_Opgave 7_**  
Bekijk de summary van het model dat je bij Opgave 6 hebt gefit.  
a. Wat is de vergelijking van het gefitte ARIMA-model?  
b. Zijn alle parameters van het model significant in de populatie? Zo nee, welke niet?  
c. Zijn de residuen normaal verdeeld volgens de summary? Waarom?

```python
print(fitted.summary())
```

a. $X_t = 0.478X_{t-1} - 0.443X_{t-2} - 0.998\varepsilon_{t-12}$  
b. Alle parameters zijn statistisch significant.  
c. Ja, de Jarque-Bera heeft een voldoende grote $p$-waarde, dus behouden we $H_0$ ($H_0$: de residuen zijn normaal verdeeld, $H_1$: de residuen zijn niet normaal verdeeld.)

Normaliteit is tot nu toe de grote ontbrekende geweest. Waar bijna alles in deze les een vergelijkbaar concept heeft bij lineaire regressie van Statistiek 2, lijkt normaliteit hier er niet echt toe te doen in vergelijking met lineaire regressie. En dat klopt: normaliteit van residuen is bij ARIMA geen vereiste. De reden hiervoor is, wederom, de wijze waarop het model gefit wordt. Een lineair regressiemodel wordt in principe met n formule gefit; er is niet meer nodig dan een eenvoudige matrixvermenigvuldiging om tot het gefitte model te komen. Bij ARIMA ligt dat anders: het model wordt daar continu bijgesteld met elk datapunt dat erbij komt (dat zie je ook aan de residuen; aan het begin bijzonder slecht, maar later worden ze steeds beter). Dat proces maakt ARIMA robuust tegen niet-normaliteit.
